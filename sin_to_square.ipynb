```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def get_fourier_component(x, n, L):
  y = (1/n)*np.sin(n*2*np.pi*x/L)
  return y

def get_reconstructed_signal (x, n, L):
  full_series = 0
  for component in range(n):
    # print(component)
    term = get_fourier_component(x,2*component+1,L)
    full_series = full_series+term
  return full_series

def make_fourier_plot(n, L):
  s=0.1
  plt.scatter(x,get_reconstructed_signal(x,1,L),color='b', s=s, label='1')
  plt.scatter(x,get_reconstructed_signal(x,n,L),color='r', s=s, label=n)
  plt.title("Sulochana's first matplotlib chart")
  plt.xlabel("Time")
  plt.ylabel("Amplitude")
  plt.legend()
  plt.grid()
  plt.show()

def make_fourier_plot2 (ax, n, L):
  s=0.1
  # ax.scatter(x,get_reconstructed_signal(x,1,L),color='b', s=s, label='1')
  ax.scatter(x,get_reconstructed_signal(x,n,L),color='r', s=s, label=n)
  ax.set_title("Number of fourier components = "+str(n))
  ax.set_xlabel("Time")
  ax.set_ylabel("Amplitude")
  ax.legend()
  ax.grid()
  plt.tight_layout()


```

```
x = np.arange(0,100,0.01)

def make_plot_array (n, L):
  fig, ax = plt.subplots(n, figsize=(12,8))
  for items in range(1, n+1, 1):
    make_fourier_plot2(ax[items-1],items,L)

make_plot_array(4,10)
```
